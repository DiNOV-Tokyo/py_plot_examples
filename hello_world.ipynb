```python
import matplotlib.pyplot as plt

month = ['Jan', 'Feb', 'Mar', 'Apr']
tmp_ave = [3.2, 6.3, 9.3, 5.0]
flg=10

plt.figure(figsize = (8,6))
plt.ylim(0, 12)
plt.grid(axis='y')
plt.grid(which='minor', axis='y', alpha= 0.5)
plt.minorticks_on()
plt.yticks(range(0,12,1))
plt.suptitle("Temperature in Japan", x=0.2, y=0.92, size=16, weight=2, color="green")
plt.legend(loc='upper left',fontsize=flg,title='Temperature')
plt.plot(month, tmp_ave)
plt.show
```

```python
plt.bar(month, tmp_ave, color = 'lightgreen', width = 0.6)
plt.show
```

```python
import matplotlib.pyplot as plt

x = [120, 200, 310, 122, 302, 40]
labels = ["A", "B", "C", "D", "E", "F"]
colors = ['red', 'pink', 'orange', 'yellow', 'lightgreen', 'green' ]

plt.pie(x, explode=[0, 0, 0.2, 0, 0, 0], labels=labels, colors=colors, autopct='%.lf%%', startangle=90, counterclock=False)
plt.show()
```

```python
y = [100, 200, 321, 234, 543, 702]

plt.scatter(x, y, c='grey')
plt.show()
```
